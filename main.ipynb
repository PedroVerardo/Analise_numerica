## Generate random points

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 5

arr = np.random.rand(num_points, 2)
arr = arr*10
arr
```

```python
x = arr[:, 0]
y = arr[:, 1]

plt.scatter(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, y, color='red', linewidth=1)


plt.show()
```

```python
#creating a closed cubic spline
arr2 = arr.copy()
arr2[num_points - 1] = arr2[0]
arr2
```

```python
xclosed = arr2[:, 0]
yclosed = arr2[:, 1]

plt.scatter(xclosed, yclosed)
plt.plot(xclosed, yclosed, color='red', linewidth=1)


plt.show()
```
